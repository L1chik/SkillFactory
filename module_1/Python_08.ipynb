```python
def profession_code(s):
    if s == "Рабочий":
        return 0
    elif s == "Менеджер":
        return 1
    else:
        return 2
profession_code("Менеджер")
sample.Profession = sample.Profession.apply(profession_code)
sample2 = sample
```

```python
def age_category(age):
    if age < 23:
        return "молодой"
    elif 23 < age < 35:
        return "средний"
    elif age > 34:
        return "зрелый"
age_category(30)
```

```python
sample['Age_category'] = sample.Age.apply(age_category)
sample
```
